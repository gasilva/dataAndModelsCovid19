```python
import unicodedata

def strip_accents(text):

    try:
        text = unicode(text, 'utf-8')
    except NameError: # unicode is a default on python 3 
        pass

    text = unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")

    return str(text)

def findMissing(a, b, n, m): 
      
    # Store all elements of second  
    # array in a hash table 
    s = dict()
    result=[]
    for i in range(m): 
        s[b[i]] = 1
  
    # Print all elements of first array  
    # that are not present in hash table 
    for i in range(n): 
        if a[i] not in s.keys(): 
            result.append(a[i])
    return result

for i in range(0,len(finalDRS)):
    finalDRS[i] = strip_accents(finalDRS[i])

dictionary = {'Ç':'C'}
data.Nome.replace(dictionary, regex=True, inplace=True)
data.Nome.replace('CENTRAL','ARARAQUARA', inplace=True)
data.Nome.replace('SAO JOSE DOS CAMPOS','TAUBATE', inplace=True)
data.Nome.replace('CAMPINAS','CAMPINAS Grande', inplace=True)


a = [x.upper() for x in finalDRS] 
b = data.Nome
n = len(a) 
m = len(b) 
newRegions=findMissing(a, b, n, m)
newRegions
```

```python
def value(r): 
    if (r == 'I'): 
        return 1
    if (r == 'V'): 
        return 5
    if (r == 'X'): 
        return 10
    if (r == 'L'): 
        return 50
    if (r == 'C'): 
        return 100
    if (r == 'D'): 
        return 500
    if (r == 'M'): 
        return 1000
    return -1
  
def romanToDecimal(str): 
    res = 0
    i = 0
  
    while (i < len(str)): 
  
        # Getting value of symbol s[i] 
        s1 = value(str[i]) 
  
        if (i + 1 < len(str)): 
  
            # Getting value of symbol s[i + 1] 
            s2 = value(str[i + 1]) 
  
            # Comparing both values 
            if (s1 >= s2): 
  
                # Value of current symbol is greater 
                # or equal to the next symbol 
                res = res + s1 
                i = i + 1
            else: 
  
                # Value of current symbol is greater 
                # or equal to the next symbol 
                res = res + s2 - s1 
                i = i + 2
        else: 
            res = res + s1 
            i = i + 1
  
    return res 
```

```python
for i in range(0,len(dataSaude)):
    numeroDR=dataSaude.iloc[i,0].split(' ',)[1]
    nomeDR=dataSaude.iloc[i,0].split(' ',)[3:]
    nomeDR=' '.join(nomeDR)
    decimalDR=romanToDecimal(numeroDR)
    if decimalDR<10:
        decimalDR="0"+str(decimalDR)
    dataSaude.iloc[i,0]="DR "+str(decimalDR)+" - "+nomeDR
```
